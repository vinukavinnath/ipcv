```python
import cv2 as cv                 #Imports opencv-python
import numpy as np               #Imports numpy
import easyocr                   #Imports easyocr
import matplotlib.image as mpimg #Imports image module from matplotlib
import matplotlib.pyplot as plt  #Imports pyplot module from matplotlib
```

```python
#Loads the image
image=mpimg.imread("./images/plate_1.jpg")
```

```python
#Displays the image
plt.imshow(image)
```

```python
#Converts image into grayscale
gray_image=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
```

```python
#Display grayscale image
plt.imshow(gray_image)
```

```python
#Applying Gaussian Blur
blurred_image=cv.GaussianBlur(gray_image,(5,5),0)
```

```python
#Display blurred image
plt.imshow(blurred_image)
```

```python
#Defines a reader to identify characters by easyocr
reader = easyocr.Reader(['en'])
```

```python
#Detects text in the numberplate and returns it
def perform_ocr(img, reader):
    # perform OCR on the preprocessed image
    result = reader.readtext(img)

    # extract and return the recognized text
    recognized_text = ' '.join([entry[1] for entry in result])
    return recognized_text
```

```python
text=perform_ocr(blurred_image,reader)
```

```python
#Displays the image and extracted text
def display_image_with_text(img, text):
    plt.imshow(img)
    plt.title(f"Recognized Text: {text}")
    plt.axis("off")
    plt.show()
```

```python
display_image_with_text(image,text)
```
