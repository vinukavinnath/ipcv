```python
#Load the image to a variable called image
image=cv.imread('./images/shapes.jpg')
```

```python
plt.imshow(image)
```

```python
#Convert color channel from BGR to RGB
rgb_image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
```

```python
#Display converted image
plt.imshow(rgb_image)
```

```python
#Converting into grayscale
gray_image=cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
```

```python
#Displaying grayscale image in Gray colormap
plt.imshow(gray_image,cmap='gray')
```

```python
#Apply Gaussian blur
blurred_image=cv.GaussianBlur(gray_image,(5, 5),0)
```

```python
#Display Gaussian blur applied image
plt.imshow(blurred_image)
```

```python
#Perform Edge Detection
edges_image=cv.Canny(blurred_image,50,150)
```

```python
#Display Edge detected image
plt.imshow(edges_image)
```

```python
#Threshold the Grayscale image
_,thresh=cv.threshold(gray_image,225,255,cv.THRESH_BINARY)
```

```python
#Display Thresholded image
plt.imshow(thresh)
```

```python
#Finding contours in edge detected image
contours, _ = cv.findContours(edges_image, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
```

```python
print(contours)
```

```python
#Takes a copy of original image
image_copy=image.copy()
```

```python
#Draw coutours on the copied image
cv.drawContours(image_copy,contours,-1,(0,0,255),5)
```

```python
#Display contour drawn image
plt.imshow(image_copy)
```

```python
# iterate through the contours and identifies polygonal curves
for contour in contours:
    # approximate the contour to a polygon
    epsilon = 0.04 * cv.arcLength(contour, True)
    approx = cv.approxPolyDP(contour, epsilon, True)

    #If polygonal curve count equivalent to 3, then draw bounding boxes
    if len(approx) == 3:
        # draw a bounding box around the triangle (in red)
        x, y, w, h = cv.boundingRect(contour)
        cv.rectangle(rgb_image, (x, y), (x + w, y + h), (0, 0, 255), 2)
```
